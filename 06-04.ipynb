```python
import time                         # measuring time
import random                       # generating arrays of random numbers
import matplotlib.pyplot as plt     # for plotting
```

```python
# Basic Introduction of matplotlib.pyplot

# To start drawing a plot, first we should start a new figure
plt.figure ()

# Can plot points using plt.plot and providing a single list/array of y-coordinates
# (the x-coordinates will simply be the indices)
plt.plot ([60, 42, 90, 7])

# To draw another plot in the same figure, we can simply use plt.plot again
# (without starting a new figure)
# when two arguments are provided, the first is for x-coordinates and the
# second is for y-coordinates
# x-coordinates do not necessarily need to be sorted
plt.plot ([1.5, 5, 0], [81, 77, 13])

plt.xlabel ("x-coordinates")              # Label for the x-axis
plt.ylabel ("y-coordinates")              # Label for the y-axis
plt.title ("Basic pyplot demonstration")  # Figure Title
plt.grid (True)                           # Show gridlines

# If you want another figure, use plt.figure () again
# Each new plot would be drawn on the latest figure

plt.figure (figsize = (4, 3))   # can adjust the size of the figure as well
plt.plot ([4000, 7421, 9801], [300, 420, 816], label = "First")
plt.plot ([5231, 7000, 20000], [640, 1400, 200], label = "Second")
plt.plot ([3333, 6666], [700, 200])     # Unlabeled plots are excluded in legend

plt.legend (loc="best")     # Adds a legend to identify labeled plots

# Can adjust the ranges for the axes
plt.xlim ([0, 30000])
plt.ylim ([0, 1500])
# plt.ylim ([0, 1000])      # axis limits may even exclude some points
```

```python
def linear1 (lst):
  cnt = 0
  for elem in lst:
    cnt += 1
  return cnt
```

```python
def randarrgen (size, minval = 1, maxval = 10000):
  lst = []
  for _ in range (size):
    r = random.randint (minval, maxval)   # produce a random number from minval to maxval inclusive
    lst.append (r)
  return lst

print (randarrgen (10))
print (randarrgen (10))
print (randarrgen (10))
print (randarrgen (10))
print (randarrgen (10))
print (randarrgen (10, 100, 999))
```

```python
# We want to measure the execution time for linear1 for sizes
# 10000, 11000, 12000, 13000, ..., 100000

startsize = 10000
endsize = 100000
stepsize = 1000

xcoord = []     # Input Sizes
ycoord = []     # Execution Time

for size in range (startsize, endsize, stepsize):
  lst = randarrgen (size)

  starttime = time.time()
  linear1 (lst)
  endtime = time.time()

  xcoord.append (len (lst))
  ycoord.append (endtime - starttime)

  #print ("Size:", len (lst))
  #print ("Execution Time:", endtime - starttime, "seconds")

plt.figure ()
plt.plot (xcoord, ycoord)

plt.xlabel ("Input Size")
plt.ylabel ("Execution Time (seconds)")
plt.title ("linear1 function time measurements")
plt.grid (True)
```
