```python
import time
```

```python
start = time.time ()

N = 1000000   # 10^6

cnt = 0
for i in range (N):
  cnt += 1
print (cnt)     # 100

end = time.time ()

print (end - start, "seconds")
```

```python
start = time.time ()

N = 10000000   # 10^7

cnt = 0
for i in range (N):
  cnt += 1
print (cnt)     # 100

end = time.time ()

print (end - start, "seconds")
```

```python
start = time.time ()

N = 20000000   # 2 * 10^7

cnt = 0
for i in range (N):
  cnt += 1
print (cnt)     # 100

end = time.time ()

print (end - start, "seconds")
```

```python
start = time.time ()

N = 100000000   # 10^8

cnt = 0
for i in range (N):
  cnt += 1
print (cnt)     # 100

end = time.time ()

print (end - start, "seconds")
```

In the examples above, the runtime was **linear** to $N$, i.e., the time was roughly proportional to the value of $N$, e.g., if you double the value of $N$, the execution time roughly doubles.

```python
start = time.time ()

N = 100000000
print (N ** 2)

end = time.time ()

print (end - start, "seconds")
```

For the single example above, the runtime is **constant** with respect to $N$, i.e., the execution time doesn't change much when we change $N$.

```python
start = time.time ()

N = 1000    # 10^3

cnt = 0
for i in range (N):
  for j in range (N):
    cnt += 1
print (cnt)     # 100? 10000?

end = time.time ()

print (end - start, "seconds")
```

```python
start = time.time ()

N = 2000    # 2 x 10^3

cnt = 0
for i in range (N):
  for j in range (N):
    cnt += 1
print (cnt)

end = time.time ()

print (end - start, "seconds")
```

```python
start = time.time ()

N = 10000    # 10^4

cnt = 0
for i in range (N):
  for j in range (N):
    cnt += 1
print (cnt)

end = time.time ()

print (end - start, "seconds")
```

In the three examples above, the runtime is **quadratic** to $N$, where the changes to $N$ are amplified by squaring, i.e., doubling the value of $N$ will cause the execution time to increase by roughly $2^2 = 4$ times.

```python
N = 100

cnt = 0
for i in range (N):
  cnt += i
print (cnt)     # sum of 0 + 1 + 2 + ... + 99? N(N + 1)/2
```
