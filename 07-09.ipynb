Proof for $x^{\log_y (z)} = z^{\log_y (x)}$

Assume $x^{\log_y (z)} = w$

If we apply $\log_y$ to both sides, we get

$\log_y \left(x^{\log_y z}\right) = \log_y (w)$

$\implies \log_y (z) \log_y (x) = \log_y (w)$

$\implies \log_y \left(z^{\log_y (x)}\right) = \log_y (w)$

$\implies z^{\log_y (x)} = w = x^{\log_y (z)}$
