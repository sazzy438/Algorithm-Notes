We also have notations to convey greater than, less than, etc with respect to growth rate.

*  We write $f(n) \in \Theta (g(n))$ to convey that the growth rate $f(n)$ is **equal to** the growth rate of $g(n)$
*  We write $f(n) \in O(g(n))$ to convey that the growth rate $f(n)$ is **less than or equal to** the growth rate of $g(n)$
*  We write $f(n) \in \Omega (g(n))$ to convey that the growth rate $f(n)$ is **greater than or equal to** the growth rate of $g(n)$
*  We write $f(n) \in o(g(n))$ to convey that the growth rate $f(n)$ is **strictly less than** the growth rate of $g(n)$ [OPTIONAL]
*  We write $f(n) \in \omega(g(n))$ to convey that the growth rate $f(n)$ is **strictly greater than** the growth rate of $g(n)$ [OPTIONAL]

We want a formal mathematical definition for these order notations.

Let's try to construct a formal definition for the $O()$ notation.

**Attempt 1:** The statement $f(n) \in O(g(n))$ means that $f(n) \leq g(n)$ for all $n$.

**Problem:** For example, we expect $200n \in O(n^2)$ but $200n \not\leq n^2$ for small values like $n = 100$.

**Suggested Fix:** Focus on large values of $n$. But how large is "large"?

**Attempt 2:** The statement $f(n) \in O(g(n))$ means that $f(n) \leq g(n)$ for all $n \geq 1000$.

**Problem:** For example, we expect $1000000n \in O(n^2)$, but $1000000n \not\leq n^2$ for values like $n = 2000$.

**Suggested Fix:** We still want to focus on large values of $n$, i.e., $n$ need to be above a certain number. But the definition should not restrict what this number can be.

**Attempt 3:** The statement $f(n) \in O(g(n))$ means that there exists a constant $n_0 > 0$ such that $f(n) \leq g(n)$ for all $n \geq n_0$.

This fixes all the previous issues that we observed. And in fact, whenever $f(n)$ has a lower growth rate than $g(n)$, this definition would actually hold. However, it may fail when the growth rates are equal.

**Problem:** We expect $12n \in O(3n)$, but $12n \not\leq 3n$ regardless of how high $n$ is.

**Suggested Fix:** We need to somehow ignore the constant factors. We can do this by allowing one side to be multiplied by a constant factor in order to achieve the desired result.

**Attempt 4 (Final Correct Definition):** The statement $f(n) \in O(g(n))$ means that there exists two constants $n_0 > 0$ and $c > 0$ such that $f(n) \leq cg(n)$ for all $n \geq n_0$.

We can define $\Omega ()$ symmetrically as well.

**Final Correct Definition:** The statement $f(n) \in \Omega(g(n))$ means that there exists two constants $n_0 > 0$ and $c > 0$ such that $f(n) \geq cg(n)$ for all $n \geq n_0$.

How do we define $\Theta ()$? One simple option is to just use $O()$ and $\Omega()$.

The statement $f(n) \in \Theta (g(n))$ means that $f(n) \in O(g(n))$ AND $f(n) \in \Omega (g(n))$.
