Given a **directed** graph, a topological ordering is a sequence of ALL vertices arranged such that every edge is from an earlier vertex to a later vertex.

image.png

A graph may have multiple topological orderings.

If a cycle exists, then the graph has no topological ordering.

If a cycle does not exist, there is always at least one topological ordering.

One of the algorithms for topological sorting involves DFS.

We pick any unmarked vertex and perform DFS. Whenever we complete the DFS of a vertex (including recursive calls), we add it to the left side of our output sequence (initially empty). When we complete the DFS for the vertex we started with, we look for another unmarked vertex to perform DFS on. Algorithm ends when all vertices are marked.

```python
# [OPTIONAL] convert index to uppercase letter
def idx2char (idx):
  return chr (ord ('A') + idx)
```

```python
numVertices = 14

A, B, C, D, E, F, G, H, I, J, K, L, M, N = range (14) # A = 0, B = 1, C = 2, ...

adjlst = [[B],           # A
          [M],  # B
          [B],           # C
          [],           # D
          [G],              # E
          [D],           # F
          [H, C, L],     # G
          [],              # H
          [F],              # I
          [A],           # J
          [L, B],           # K
          [J, B, M],        # L
          [],           # M
          [D]]              # N

visited = [False] * numVertices
topo = []

def topodfs (adjlst, curr):
  visited[curr] = True
  for nbr in adjlst[curr]:
    if not visited[nbr]:
      topodfs (adjlst, nbr)
  topo.append (curr)

for u in range (numVertices):
  if not visited[u]:
    topodfs (adjlst, u)
topo.reverse ()

for elem in topo:
  print (idx2char (elem), end = ' ')
```

Runtime is $\Theta (|V| + |E|)$ for all cases
