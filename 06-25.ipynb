```python
def printfirstlast (lst):   # 2 steps
  print (lst[0])
  print (lst[-1])

# Runtime is in Theta (1)
```

```python
def printall (lst):
  for elem in lst:
    print (elem)

# Runtime is in Theta (n)
```

```python
def printplusfive (lst):
  for elem in lst:
    print (elem + 5)

# Runtime is in Theta (n)
```

```python
def combine (lst):
  for elem1 in lst:           # this loop runs n times
    for elem2 in lst:         # for each iteration of the first loop, this loop runs n times
      print (elem1 + elem2)   # this runs a total of n*n = n^2 times

# Runtime is in Theta (n^2)
```

**Example 7 from Lecture 2**

The value of $s$ after $i$ iterations is $s = 1 + 2 + 3 + \cdots + i$

We want to know the runtime, which is based on how many times the loop runs.

Suppose the loop runs $t$ times.

The loop ends when the value of $s$ crosses the value of $n$. But what is the value of $s$ at this time?

After $t$ iterations, the value of $s = 1 + 2 + 3 + \cdots + t = \frac{t(t + 1)}{2}$

The loop ends when the value of $s$, which is of order $\Theta (t^2)$ reaches $n$. This implies that the value of $t$ reaches the order of $\Theta (\sqrt{n})$. Without needing to calculate the exact function, we can conclude that the loop ends when $t$ is some expression whose growth rate is in $\Theta (\sqrt{n})$, i.e., the runtime is in $\Theta (\sqrt{n})$.
